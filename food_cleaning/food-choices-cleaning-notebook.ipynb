Afther further inspection I noticed that most of the data in this dataframe is collected in an ordinal fashion. Therefore I will aim to change the missing values according to the ordinal values provided in the docx document and remove columns that might be considered duplicate data.

```python
drop_cols = ['type_sports', 'mother_profession', 'meals_dinner_friend',
            'ideal_diet', 'healthy_meal', 'food_childhood', 'fav_cuisine',
            'father_profession', 'eating_changes', 'diet_current', 'comfort_food',
            'comfort_food_reasons']
```

Now after selecting which columns we do not need in our dataframe we are going to check for missing values in the object datatype columns

```python
((df.drop(columns=drop_cols)
  .select_dtypes('object')
  .isna()
  .sum()
))
```

Now that we have reduced the number of missing values in the dataframe by removing some of the object data type columns that had duplicate data, I am ready to fill the missing values in the float data type columns.

The docx document 'codebook_food.docx' contains the ordinal values for the remaining columns in the dataframe. Following the indicated ordinal values
in the docx document I will fill the missing values as if I was using an imputter that fills the missing values with the most common value

```python
df['calories_scone'] = df['calories_scone'].fillna(*df['calories_scone'].mode())
df['father_education'] = df['father_education'].fillna(*df['father_education'].mode())
df['income'] = df['income'].fillna(*df['income'].mode())
df['life_rewarding'] = df['life_rewarding'].fillna(*df['life_rewarding'].mode())
df['mother_education'] = df['mother_education'].fillna(*df['mother_education'].mode())
df['on_off_campus'] = df['on_off_campus'].fillna(*df['on_off_campus'].mode())
df['self_perception_weight'] = df['self_perception_weight'].fillna(*df['self_perception_weight'].mode())
df['soup'] = df['soup'].fillna(*df['soup'].mode())
df['drink'] = df['drink'].fillna(*df['drink'].mode())

```

For the rest of the float type columns there are ordinal options in the docx document that point to value for missing value, therefore I will use the value provided by the docx document to fill the missing values.

```python
df['comfort_food_reasons'] = df['comfort_food_reasons'].fillna(9)
df['cook'] = df['cook'].fillna(5)
df['cuisine'] = df['cuisine'].fillna(6)
df['employment'] = df['employment'].fillna(4)
df['exercise'] = df['exercise'].fillna(5)
df['marital_status'] = df['marital_status'].fillna(1)
df['sports'] = df['sports'].fillna(99)
# Additionally removing 'fav_cuisine' since 'fac_cuisine_coded' already exist
drop_cols.append('fav_cuisine')
```

As mentioned before, most of the values in these columns are ordinally encoded. However, the datatype for these columns appear to be float datatype
therefore we need to change the data type back to a int since its an ordinally encoded dataframe.

```python
# collecting float number columns to convert to int16
int_col = df.select_dtypes('float64').columns
```

Lastly we need to make some additional chnages to the last two object type columns and fill their missing values. Looking closely to the data in both of these columns we can see that there may be come values that were entered incorrectly into the dataframe. To fix this issue we will grab all of the 
numerical values with decimal places in the GPA column and all the numerical values in the wieght column.

```python
df['GPA'] = df['GPA'].str.extract(pat='(\d+\.\d+)', expand=True)
df['weight'] = df['weight'].str.extract(pat='(\d+)', expand=True)
```

Lastly I will create a dictionary to set the float data type columns into int data types with the exception of the GPA column which has float values

```python
col_dict = dict(set((i, 'int16') for i in int_col))
col_dict['GPA'] = 'float32'
col_dict['weight'] = 'int16'

((df
  .drop(columns=drop_cols)
  .dropna()
  .astype(col_dict)
 ))
```
